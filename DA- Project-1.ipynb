```python
# 4) What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this
```

```python
geme=df.groupby(['brand','vehicleType'])['price'].mean().unstack()
geme
```

```python
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 8))
sns.heatmap(geme, cmap="YlGnBu", annot=True, fmt=".0f", cbar_kws={'label': 'Average Price'})
plt.title("Average Price of Vehicles by Brand and Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Brand")
plt.show()
```
